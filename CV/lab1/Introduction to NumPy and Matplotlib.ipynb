### Operations on arrays

Say you want to add one to each number in an array.  We could do this with a for loop but you probably know by now that we **do not** want to do it that way!!

```python
c = np.array([1,2,3])
# do not do this
for i in range(3):
    c[i] += 1
c
```

With NumPy you can naturally perform arithmetic on arrays.  The operation will automatically be applied to every element in the array.

```python
c = np.array([1,2,3]) # because it is a vector you can add to the vector and it applies to all elements
c += 1
c
```

NumPy also provides lots of *vectorized* math functions.  *Vectorized* means that the operation will be applied to every element in the array, just as above.

```python
theta = np.array([0,np.pi/4,np.pi/2])
np.sin(theta)
```

We can also apply arithmetic on arrays in a natural way.

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
a + b
```

### Iterating over arrays

If you iterate over an array, it will iterate over the first dimension.  So if you iterate over a 2D matrix, you will iterate over the rows.

```python
a = np.array([[1,2,3],[4,5,6]])
for row in a:
    print(row)
```

If you provide NumPy with a list, it will implicitly convert it into an NumPy array.  You can even provide it a list of arrays.  Lists (and lists of lists, etc.) are pretty much interchangeable with arrays.

```python
a = [[1,2,3],[4,5,6]]
np.sin(a)
```

### Combining arrays

You can use `concatenate` and `stack` to combine arrays.  `concatenate` combines along an existing dimension (or axis in NumPy lingo), and `stack` creates a new axis.

Note that axis `-1` refers to the last dimension.

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
```

```python
c = np.concatenate([a,b],axis=0)
c
```

```python
c.shape
```

```python
c = np.stack([a,b],axis=1)
c
```

```python
c.shape
```

```python
c = np.stack([a,b],axis=-1) #same as 1??
```

```python
c
```

```python
c.shape
```

### Beyond matrices

NumPy can handle more than 2 dimensions/axes.  

```python
a = np.array([[1,2,3],[4,5,6]])
np.stack([a,a,a],axis=-1)
```

```python
np.zeros((8,8,3))
```

### Useful NumPy functions

`linspace` will make a list of equally-spaced numbers between a start and end points.  `endpoint=True` tells NumPy to include the endpoint in the list.

```python
x = np.linspace(0,10,50,endpoint=True) # 50 values between the space of 0,10, 
x
```

```python
y = np.sin(x)
y
```

## Matplotlib

Matplotlib is a Python library for making plots and showing images.

The basic command is `plt.plot(x,y)` which will make a 2D plot.

`plt.show()` is not really necessary in a notebook, but in a Python script it would make the window with the plot actually appear.

```python
x = np.linspace(0,10,50,endpoint=True)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```
