2. Make 1D edge detection filters, run them on the `love_and_peace.jpg` image, and show the result.

This filter detects vertical edges:

$\begin{bmatrix}-1 & 1\end{bmatrix}$

and this filter detects horizontal edges:

$\begin{bmatrix}-1 \\ 1\end{bmatrix}$


```python

import imageio
import numpy as np
from matplotlib import pyplot as plt
import skimage
from scipy.ndimage import convolve
import scipy
```

```python
vertical_filter = np.array([[-1,1]])

horizontal_filter = np.array([[-1],[1]])
im = imageio.imread('love_and_peace.jpg') # image is 3x3 so do I need 3x3 filter
im = skimage.color.rgb2gray(im)
plt.imshow(im, cmap = 'gray')
```

```python
im_edge = convolve(im, vertical_filter)
plt.imshow(im_edge, cmap='gray')
plt.show()
```

```python
im_edge_h = convolve(im, horizontal_filter)
plt.imshow(im_edge_h, cmap='grey')
plt.show()
```

2. A Difference of Gaussians (DoG) filter is another edge detection filter (approximation of a Laplacian filter.)

To compute the DoG filter, make two Gaussian filters with the same shape but different sigma (std. dev.) values, and then compute the difference of the two filter matrices.

*Note: This should be a 2D filter, not 1D.*

*Note: Make sure to normalize the Gaussians (divide each Gaussian by its sum) before computing the difference.*

Apply your DoG filter to the flower image and show the result.



```python
input = imageio.imread('love_and_peace.jpg')
input = skimage.color.rgb2gray(input)
G1 = scipy.signal.windows.gaussian(9, std = 1)
G2 = scipy.signal.windows.gaussian(9, std = .5)
G1_filter = (G1[:,None] @ G1[None,:] ) /81
G2_filter = (G2[:, None] @ G2[None,:] ) / 81

DoG = G1_filter - G2_filter
input_blue = convolve(input, DoG)
plt.imshow(input_blue, cmap = 'gray')
plt.axis('off')
plt.show()
```
